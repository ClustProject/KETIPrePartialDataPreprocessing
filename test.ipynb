```python
def HampelDetection(data, window_size=7, n_sigma=3):
    n = len(data)
    new_data = data.copy()
    new_data['index']=list(range(len(new_data.index)))
    new_data = new_data.set_index('index')

    k = 1.4826 # scale factor for Gaussian distribution
    
    for column in data.columns:
        for i in range((window_size),(n - window_size)):
            x0 = np.nanmedian(new_data[(i - window_size):(i + window_size)])
            S0 = k * np.nanmedian(np.abs(new_data[(i - window_size):(i + window_size)] - x0))
            if (np.abs(new_data[i] - x0) > n_sigma * S0):
                new_data[i] = np.nan
        data[column] = new_data[column]
    return data

```
